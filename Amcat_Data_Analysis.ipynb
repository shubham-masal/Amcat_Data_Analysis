```python
categorical_columns = ['CollegeTier', 'English', 'Logical', 'Quant', 'ComputerProgramming', 
                       'ElectronicsAndSemicon', 'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 
                       'TelecomEngg', 'CivilEngg', 'DOJ', 'DOL', 'Designation', 'JobCity', 'Gender', 
                       'Degree', 'Specialization']

columns_to_plot = [col for col in categorical_columns if col in df.columns]

num_plots = len(columns_to_plot)

rows = (num_plots // 3) + 1 if num_plots % 3 != 0 else num_plots // 3
cols = 3

fig, axes = plt.subplots(rows, cols, figsize=(18, rows * 5))
axes = axes.flatten()

for i, col in enumerate(columns_to_plot):
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].tick_params(axis='x', rotation=45) 

for j in range(i+1, rows*cols):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# Visualization For Bivariate

### Con Vs Con

```python
sns.set(style='whitegrid')
continuous_features = [
    '10percentage', '12percentage', 'collegeGPA', 'English', 'Logical', 
    'Quant', 'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon', 
    'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 
    'CivilEngg',  'agreeableness', 'extraversion', 
    'nueroticism', 'openess_to_experience'
]

num_features = len(continuous_features)
ncols = 3
nrows = (num_features + ncols - 1) // ncols  

plt.figure(figsize=(20, 5 * nrows))

for i, feature in enumerate(continuous_features):
    plt.subplot(nrows, ncols, i + 1)
    sns.scatterplot(data=df, x=feature, y='Salary', hue='Salary', palette='viridis', s=100)
    plt.title(f'Scatter Plot: {feature} vs. Salary')
    plt.xlabel(feature)
    plt.ylabel('Salary')
    plt.grid()

plt.tight_layout()
plt.show()
```

###  Insights

* These plots showing there is no strong connection between con vs con columns
* There are some few columns who show a single line patterns
* rest other columns are randomly spread

```python
sns.set(style="whitegrid")

categorical_columns = ['Gender', 'Degree', 'Specialization']

plt.figure(figsize=(20, 15))  

for i, col in enumerate(categorical_columns, 1):
    plt.subplot(3, 2, i)  
    sns.boxplot(x=col, y='Salary', data=df)
    plt.title(f'Box Plot of Salary vs {col}')
    plt.xticks(rotation=45)  

plt.tight_layout()  
plt.show()  
```

# Cat vs Cat

```python
sns.set(style="whitegrid")

categorical_columns = ['Gender', 'Degree', 'Specialization']

plt.figure(figsize=(20, 15))  

for i, col1 in enumerate(categorical_columns):
    for j, col2 in enumerate(categorical_columns):
        if i < j:  
            plt.subplot(len(categorical_columns)-1, len(categorical_columns)-1, i*(len(categorical_columns)-1) + j)  # Grid layout for subplots
            sns.countplot(data=df, x=col1, hue=col2, palette="Set2")
            plt.title(f'Grouped Bar Plot: {col1} vs {col2}')
            plt.xticks(rotation=45)  

plt.tight_layout() 
plt.show() 
```

### Insights

* It showing the males who have done btech they actully have more number of count
* female has least count in mca
* there is no one who has done Mtech/ME
